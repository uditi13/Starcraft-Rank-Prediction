```python
# For Normal Distribution the skewness level should be between -1 and 1

for column in df2.columns:
    print(column,' ', df2[column].skew())
```

```python
outlier_cols = []
for column in df2.columns:
    if (df2[column].skew()>2) or (df2[column].skew()<-2):
        outlier_cols.append(column)
        
print(outlier_cols)
```

```python
# function to generate box_plots for the outlier columns
def plot_boxplot(num_plots):
    num_rows = (num_plots // 3) + (num_plots % 3 > 0)
    num_cols = min(num_plots, 3)

    fig, axes = plt.subplots(1, num_plots, figsize=(12, 6))

    if num_rows == 1 or num_cols == 1:
        axes = axes.flatten()

    for i, column in enumerate(outlier_cols[:num_plots]):
        sns.boxplot(data=df2, x=column, ax=axes[i])
        axes[i].set_title(column)

    plt.tight_layout()

    plt.show()
```

```python
# Plotting boxplots before fixing outliers
plot_boxplot(len(outlier_cols[:4]))
plot_boxplot(len(outlier_cols[4:]))
```

```python
def outlier_fixation(column):
    Q1 = df2[column].quantile(0.25)
    Q3 = df2[column].quantile(0.75)
    IQR = Q3 - Q1
    whisker_width = 1.5
    lower_whisker = Q1 -(whisker_width*IQR)
    upper_whisker = Q3 + (whisker_width*IQR)
    df2[column]=np.where(df2[column]>upper_whisker,upper_whisker,np.where(df2[column]<lower_whisker,lower_whisker,df2[column]))
```

```python
for col in outlier_cols:
    outlier_fixation(col)
```

```python
# Plotting boxplots after fixing outliers
plot_boxplot(len(outlier_cols[:4]))
plot_boxplot(len(outlier_cols[4:]))
```

```python
import pickle
import numpy as np
import pandas as pd
import sklearn as sk
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC, SVC
from sklearn.feature_selection import f_classif
from sklearn.preprocessing import label_binarize
from sklearn.decomposition import PCA as pca_skl
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
```

```python
dataset = df2
```

```python
def summary(model, X_test, y_test, display=True):
    
    y_pred = model.predict(X_test)
    MSE = np.mean((y_test - y_pred)**2)
    RMSE = np.sqrt(MSE)
    R_squared = 1 - np.sum((y_test - y_pred)**2) / np.sum((y_test - np.mean(y_test))**2)
    
    if display:
        print('Accuracy :', model.score(X_test, y_test))
        print('MSE: ', MSE)
        print('RMSE: ', RMSE)
        print('R²: ', R_squared)
    else:
        return [model.score(X_test, y_test), MSE, RMSE, R_squared]
```

```python
X = dataset.drop(noncorrelated_columns, axis=1)
X = X.drop('LeagueIndex', axis=1)
names = X.columns
X = X.to_numpy()
Y = dataset['LeagueIndex'].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, Y, 
                                                    random_state=42,
                                                    shuffle=True,
                                                    test_size=0.2, stratify = Y)
scl = sk.preprocessing.StandardScaler()
X_train_scaled = scl.fit_transform(X_train)
X_test_scaled = scl.transform(X_test)
```

```python
names
```

```python
lgr_clf = LogisticRegression(max_iter=10000)
lgr_clf.fit(X_train_scaled, y_train)
summary(lgr_clf, X_test_scaled, y_test)
```

```python
rf_clf = RandomForestClassifier(n_estimators=1000)
rf_clf.fit(X_train, y_train)
summary(rf_clf, X_test, y_test)
```

__Since the accuracy was very low, I included the rows of LeagueIndex = 8 (dropped the columns: Age, HoursPerWeek, TotalHours) and ran the entire notebook again and found that the accuracy had worsened.__

__This suggests that TotalHours is a prominent feature for predicting the LeagueIndex (Rank). This also justifies dropping all those rows that had '?' as these rows were lacking an important feature for classification.__

__Also, this dataset is very unbalanced as we observe that there is more data from a particular LeagueIndex compared to others. This implies that we would need more data apart from the columns given in thsi dataset to determine the Rank of the player accurately.__

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, Y)
```

```python
unique, counts = np.unique(y_resampled, return_counts=True)
class_distribution = dict(zip(unique, counts))
print("Class Distribution:", class_distribution)
```

```python
X_resampled_train, X_resampled_test, y_resampled_train, y_resampled_test = train_test_split(X_resampled, y_resampled, 
                                                    random_state=42,
                                                    shuffle=True,
                                                    test_size=0.2, stratify = y_resampled)
scl = sk.preprocessing.StandardScaler()
X_resampled_train_scaled = scl.fit_transform(X_resampled_train)
X_resampled_test_scaled = scl.transform(X_resampled_test)
```

```python
lgr_clf = LogisticRegression(max_iter=10000)
lgr_clf.fit(X_resampled_train_scaled, y_resampled_train)
summary(lgr_clf, X_resampled_test_scaled, y_resampled_test)
```

```python
rf_clf = RandomForestClassifier(n_estimators=1000)
rf_clf.fit(X_resampled_train_scaled, y_resampled_train)
summary(rf_clf, X_resampled_test_scaled, y_resampled_test)
```
